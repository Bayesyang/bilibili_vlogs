axis = i，在第 i 个轴（axis） 上进行操作

X.shape == (N, C, H, W), X.sum(axis=i)
- axis = 0, X[0, :, :, :] + X[1, :, :, :] + ... + X[N-1, :, :, :]
    - 作为一种降维/collapse（沿着 axis = i，reduction to 1），(N, C, H, W) => (C, H, W)
- axis = 1, X[:, 0, :, :] + X[:, 1, :, :] + ... + X[:, C-1, :, :]
    - 作为一种降维/collapse（沿着 axis = i，reduction to 1），(N, C, H, W) => (N, H, W)
   
2D 
- axis = 0， 竖切（from up to down）
- axis = 1，横切（across, from left to right）

3D
- axis = 0
- axis = 1
- axis = 2

np
- sum/max/min/mean/std
- concat/stack
- squeeze/unsqueeze

```python
# 2张图片，3通道，4行4列
# X = np.random.randint(0, 5, (2, 3, 2, 2))
X = np.arange(24).reshape(2, 3, 2, 2)
```

```python
X
```

```python
X.shape
```

```python
print(X.sum(axis=0).shape)
# 等价于
print((X[0, :, :, :] + X[1, :, :, :]).shape)
```

```python
print(X.sum(axis=1).shape)
# 等价于
print((X[:, 0, :, :] + X[:, 1, :, :] + X[:, 2, :, :]).shape)
```
