```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

$$
\begin{split}
&z=f(x,y)\\
&z-f(x,y)=0
\end{split}
$$

- $z=f(x,y)$ 是三维空间的一个面（surface）
    - ax.plot_surface(x, y, z)
- 在 matplotlib
    - y=f(x), plt.plot(x, y) 描点连线
    - z=f(x, y), ax.plot_surface(x, y, z) 描点连面
- 自变量和因变量
    - y=f(x)，x 是自变量（independent variable），y 是因变量（dependent variable）；
        - 确定了 x 就唯一地确定了唯一的 y，自由度为1
    - 对于 z = f(x, y), x, y 是自变量（independent variables），z 是因变量（dependent variable）；
        - 确定了x, y 就唯一的确定了 z，自由度为2


```python
xs, ys = np.meshgrid(np.arange(-2, 2.01, .01), np.arange(-2, 2.01, .01))
```

## $f(x,y)=\exp\left(-(x^2+y^2)\right)$

```python
zs = np.exp(-(xs**2+ys**2))
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(xs, ys, zs, cmap=cm.jet)
fig.colorbar(surface)
```

## interactive

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib notebook
```

```python
fig = plt.figure()
ax = Axes3D(fig)

# creating the plot
surface = ax.plot_surface(xs, ys, zs, color='green', cmap=plt.get_cmap('rainbow'))

# setting title and labels
ax.set_title("3D plot")
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
fig.colorbar(surface)
# displaying the plot
plt.show()
```
