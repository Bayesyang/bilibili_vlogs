$$
\begin{split}
&z=f(x,y)\\
&z-f(x,y)=0
\end{split}
$$

- $z=f(x,y)$ 是三维空间的一个面（surface）
    - ax.plot_surface(x, y, z)
- 在 matplotlib
    - y=f(x), plt.plot(x, y) 描点连线
    - z=f(x, y), ax.plot_surface(x, y, z) 描点连面

## $f(x,y)=\exp\left(-(x^2+y^2)\right)$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python

xs, ys = np.meshgrid(np.arange(-2, 2.01, .01), np.arange(-2, 2.01, .01))
```

```python
xs.shape
```

```python
ys.shape
```

```python
zs = np.exp(-(xs**2+ys**2))
```

```python
zs.shape
```

```python
fig = plt.figure(figsize=(16, 12))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(xs, ys, zs, cmap=cm.jet)
fig.colorbar(surface)
```
